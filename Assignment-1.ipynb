### Assignment 1: Hypothesis Testing

Large Sample Tests
> Suppose we want to test a set of hypotheses concentrating a parameter θ based on a random sample $Y_1, Y_2, Y_3, ..., Y_n$. Let the large sample estimators be defined as $\bar{Y}$ ( sample mean ) and $\hat{p}$ ( unbiased point estimator ). The estimators are used to compare the two population means are: ($\mu_1 - \mu_2$) and for the comparison parameters ($p_1 - p_2$).
>
> If $\theta_0$ is a number of $\theta$, we may wish to set up our hypothesis testing as: $H_0 = \theta_0$ vs  $H_a > \theta_0$.
>
> This means that our test statistic can be written as $\hat{\theta}$ our rejection region ( the region in which we will reject the null hypothesis if the test statistic falls within it ) will be written as RR = $\hat{\theta} > k$ for some choice of k.
> The actual value of k in the rejection region RR is determined by fixing the type I error probability $\alpha$ (the level of the test) and choosing k accordingly. If $H_0$ is true, $\hat{\theta}$ has an approximately normal distribution with mean $\theta_0$ and standard error, $\sigma_0$
>
>For null distributions, Z, with a standard normal distribution then $z_\alpha$ is P(Z > $z_\alpha$) = $\alpha$.
>
> Our rejection region will then be: RR = {$\hat{\theta}$ : $\hat{\theta}$ > $\theta_0$ + $z_\alpha\sigma_\hat{\theta}$} = {$\hat{\theta}$ : $\frac{\hat{\theta} - \theta_0}{\sigma_\hat{\theta}}$ > $z_\alpha$}
>
> If we set Z = ($\hat{\theta}$ - $\theta_0$)/$\sigma_\hat{\theta}$ then our rejection region can be written in simpler terms: RR = {z > $z_\alpha$}
>
> We can rewrite our hypotheses as: $H_0$ : $\theta$ = $\theta_0$ vs $H_a$ : $\theta >  \theta_0$
>
> Our test statistic can be written as: Z = $\frac{\hat{\theta} - \theta_0}{\sigma_\hat{\theta}}$
>
> *note that the sign within our rejection region changes with our alternative hypothesis*
>
> Also notice that Z = $\frac{estimator for parameter - value for the parameter given by H_0}{standard error of the estimator}$
>
> As statistical testing is not perfect, errors may occur.
>
> A Type I error is defined as an error made when we reject the null hypothesis when we are not supposed to (meaning the we reject a "true" null hypothesis) and a Type II error is defined an error made when we fail to reject a null hypothesis when we are supposed to (meaning we fail to reject a "false" null hypothesis). 
>
> Type I and Type II error can be calculated by finding the corresponding tail area associated with the test statistic.

Example 10.6 from 120c textbook, Mathematical Statistics with Applications 
> A machine in a factory does not need to be repaired if it produces less than 10% defectives among the large lot of items that it produces in a day. A random sample of n (sample size) items from the day’s production contains y defectives, and the supervisor says that the machine does not need to be repaired. Does the sample evidence support his decision?

<center>$H_0$: p = 0.10 v.s. $H_A$: p < 0.10 </center>

<center>$H_0$ = the machine in a factory produces 10% defective items </center>
    
<center> $H_A$ = the machine in a factory produces less than 10% defective items </center>

Type I error = We reject the null hypothesis and claim that the machine in a facotry produces less than 10% defectives when it the machine only produces 10% defectives.

Type II error = We accept the null hypothesis and claim that the machine only produces 10% defective items when the machine actually produces less than 10% defective items.

<center>z = $\frac{\hat{p} - p_0}{\sigma_\hat{p}}$ = $\frac{\hat{p} - p_0}{\sqrt{p_0(1 - p_0)/n}}$ ; note that our test statistic is based on $\hat{p}$ = y/n & $p_0$ = 0.1</center>


```python
import numpy as np
import math as math
import random
import matplotlib.pyplot as plt
```

```python
# coding our test statistic: 

def teststat(p0, n, t):
    # t is the number of times you want to run this simulation for this particular sample size
    s = [ ]
    for i in range (0, t):
        y = random.randint(0, n)
        phat = y/n
        standarddev = math.sqrt(p0*(1-p0)/n)
        z = (phat-p0)/standarddev
        s.append(z)
        true_stat = np.mean(s)
    return true_stat
        
```

```python
# running the test statistic function for 200 trials, 50 times each time to find the true value
t = []
for n in range(1,201):
    tstat = teststat(0.1, n, 50)
    t.append(tstat)
print (t)
```

```python
# plotting the test statistic results
plt.plot(t, color = "green")
plt.ylabel('true test statistic')
plt.show()
```

In order to calculate the type 1 and type 2 error, we must set up an arbitrary rejection region (because we are running a general simulation and not a specific problem)


Because we reject the null hypothesis when the machine produces more than 10% defective products we are going to say that a good number to reject the null hypothesis when the machine produces more than "y" defective products. We are going to set "y" = 10 so that the rejection region is RR { y $\geq $ 10 }.

** Note that we cannot keep changing the number of defectives in this scenario because the error depends on the number selected for the rejection region and if we keep changing the number of defectives the error values project an erroneous pattern. 

```python
# Calculating Type I Error: 

# create a n choose k function first:
def ncr(n, k):
    a = (math.factorial(n))/((math.factorial(k)*math.factorial(n-k)))
    return a

def type1(p0, n, t):
    # once again, t is the number of times you want to run this simulation:
    s = [] 
    for i in range(0, t):
        alpha = 0.0
        y = 10
        for j in range(0, y+1):
            alpha += float(ncr(n, j))*((p0)**j)*((1-p0)**(n-j))
        s.append(alpha)
        truetype1 = np.mean(s)
    return truetype1
```

```python
x = [ ]

for n in range (11, 201): # we have to start with a number bigger than 10 because that is the smallest number of defectives we can have for a type I error
    t1 = type1(0.1, n, 50)
    x.append(t1)
print(x)
```

```python
# plotting the type I error statistic results
t1_range = np.arange(11, 201)
plt.plot(t1_range, x, color = "purple")
plt.ylabel('test I error')
plt.show()
```

```python
# Calculating Type II Error:
# P(Type II Error) = P(Y < 15 when p = 0.1)

def type2(p0, n, t):
    for i in range(0, t):
        beta = 0.0
        y = 10
        for j in range(y, n+1):
            beta += float((ncr(n, j))*((p0)**j)*((1-p0)**(n-j)))
    return beta
```

```python
z = [ ]
for n in range(11, 201):
    t2 = type2(0.1, n, 50)
    z.append(t2)
print (z)
```

```python
# plotting the type I error statistic results
t2_range = np.arange(11, 201)
plt.plot(t2_range, z, color = "skyblue")
plt.ylabel('test II error')
plt.show()
```
